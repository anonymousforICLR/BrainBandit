```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('default')

if __name__ == '__main__':

    Agents = [ 'PSRL',  'UBE_TS', 'UBE_UCB','UCRL2', 'UBE_BBN']
    labels = [ 'PSRL',  'UBE_TS', 'UBE_UCB', 'UCRL2', 'UBE_BBN']
    plt
    for i,Agent in enumerate(Agents):
        coverage = np.load('saved/L100/coverage_rate_' + Agent + '.npy')
        coverage = coverage[:,0:200]
        total_n = coverage.shape[0]
        L = coverage.shape[1]
        coverage_mean = np.mean(coverage, axis=0)
        coverage_std = np.std(coverage, axis=0)
        plt.fill_between(np.arange(1, L+1, 1), coverage_mean + coverage_std, coverage_mean-coverage_std, alpha=0.25)
        plt.plot(coverage_mean, label=labels[i])
    plt.legend(prop={'size': 17})
    plt.xlabel('episodes', size=20)
    plt.ylabel('coverage rate', size=20)
    plt.xticks(size=15)
    plt.yticks(size=15)
    plt.show()
```
