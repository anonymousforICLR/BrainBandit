```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
plt.style.use('default')
```

```python
RU = 1
r1_std_list = [3.5,3,2.5,2]
n = len(r1_std_list)
for (index,r1_std) in enumerate(r1_std_list):
    r2_std = r1_std - RU
    p_list_combined = []
    for times in range(10):
        p_list = np.load(f'saved/simulation_RU={RU}_std={r1_std}_neutral_{times}_n=2.npy')
        p_list = p_list[np.newaxis, :]
        if times == 0:
            p_list_combined = p_list
        else:
            p_list_combined = np.concatenate([p_list_combined, p_list], axis=0)
    p_mean = np.mean(p_list_combined, axis=0)
    p_std = np.std(p_list_combined, axis=0)
    my_colors = cm.Oranges(index/n+0.3)
    plt.plot(np.arange(-1.5, 1.5, 0.05), p_mean,color=my_colors,label=f'$\sigma_1={r1_std},\sigma_2={r1_std-RU}$ ')
    plt.fill_between(np.arange(-1.5,1.5,0.05), p_mean + p_std, p_mean - p_std, color=my_colors, alpha=0.25)
plt.axvline(0, 0, 1, linestyle='--', color='black')
plt.axhline(0.5, 0, 1, linestyle='--', color='black')
plt.xlabel('value difference', size=20)
plt.ylabel('choice probability', size=20)
plt.legend(prop = {'size':12})
plt.xticks(size=20)
plt.yticks(size=20)
plt.show()

for (index,r1_std) in enumerate(r1_std_list):
    r2_std = r1_std - RU
    p_list_combined = []
    for times in range(10):
        p_list = np.load(f'saved/simulation_RU={RU}_std={r1_std}_seeking_{times}_n=2.npy')
        p_list = p_list[np.newaxis, :]
        if times == 0:
            p_list_combined = p_list
        else:
            p_list_combined = np.concatenate([p_list_combined, p_list], axis=0)
    p_mean = np.mean(p_list_combined, axis=0)
    p_std = np.std(p_list_combined, axis=0)
    my_colors = cm.Blues(index/n+0.3)
    plt.plot(np.arange(-1.5, 1.5, 0.05), p_mean,color=my_colors,label=f'$\sigma_1={r1_std},\sigma_2={r1_std-RU}$ ')
    plt.fill_between(np.arange(-1.5, 1.5, 0.05), p_mean + p_std, p_mean - p_std, color=my_colors, alpha=0.25)
plt.axvline(0, 0, 1, linestyle='--', color='black')
plt.axhline(0.5, 0, 1, linestyle='--', color='black')
plt.xlabel('value difference',size=20)
plt.ylabel('choice probability',size=20)
plt.legend(prop = {'size':12})
plt.xticks(size=20)
plt.yticks(size=20)
plt.show()

for (index,r1_std) in enumerate(r1_std_list):
    r2_std = r1_std - RU
    p_list_combined = []
    for times in range(10):
        p_list = np.load(f'saved/simulation_RU={RU}_std={r1_std}_averse_{times}_n=2.npy')
        p_list = p_list[np.newaxis, :]
        if times == 0:
            p_list_combined = p_list
        else:
            p_list_combined = np.concatenate([p_list_combined, p_list], axis=0)
    p_mean = np.mean(p_list_combined, axis=0)
    p_std = np.std(p_list_combined, axis=0)
    my_colors = cm.Greens(index/n+0.3)
    plt.plot(np.arange(-1.5, 1.5, 0.05), p_mean,color=my_colors,label=f'$\sigma_1={r1_std},\sigma_2={r1_std-RU}$ ')
    plt.fill_between(np.arange(-1.5, 1.5, 0.05), p_mean + p_std, p_mean - p_std, color=my_colors, alpha=0.25)
plt.axvline(0, 0, 1, linestyle='--', color='black')
plt.axhline(0.5, 0, 1, linestyle='--', color='black')
plt.xlabel('value difference',size=20)
plt.ylabel('choice probability',size=20)
plt.legend(prop = {'size':12})
plt.xticks(size=20)
plt.yticks(size=20)
plt.show()



TU = 2.5
r1_var_list = [2.,1.75,1.5,1.25]
n = len(r1_var_list)
for (index,r1_var) in enumerate(r1_var_list):
    r1_std = np.sqrt(r1_var)
    r2_std = np.sqrt(TU-r1_var)
    p_list_combined = []
    for times in range(10):
        p_list = np.load(f'saved/simulation_TU={TU}_var={r1_var}_averse_{times}_n=2.npy')
        p_list = p_list[np.newaxis,:]
        if times == 0:
            p_list_combined = p_list
        else :
            p_list_combined = np.concatenate([p_list_combined,p_list],axis=0)
    p_mean = np.mean(p_list_combined,axis=0)
    p_std = np.std(p_list_combined,axis=0)
    my_colors = cm.Greens(index/n+0.3)
    plt.plot(np.arange(-1.5,1.5,0.05), p_mean, color=my_colors,label=f'$\sigma_1^2={r1_var},\sigma_2^2={TU-r1_var} $')
    plt.fill_between(np.arange(-1.5,1.5,0.05), p_mean + p_std, p_mean - p_std, color=my_colors, alpha=0.25)
plt.axvline(0, 0, 1, linestyle='--', color='black')
plt.axhline(0.5, 0, 1, linestyle='--', color='black')
plt.xlabel('value difference',size=20)
plt.ylabel('choice probability',size=20)
plt.legend(prop = {'size':12})
plt.xticks(size=20)
plt.yticks(size=20)
plt.show()


for (index,r1_var) in enumerate(r1_var_list):
    r1_std = np.sqrt(r1_var)
    r2_std = np.sqrt(TU-r1_var)
    p_list_combined = []
    for times in range(10):
        p_list = np.load(f'saved/simulation_TU={TU}_var={r1_var}_neutral_{times}_n=2.npy')
        p_list = p_list[np.newaxis,:]
        if times == 0:
            p_list_combined = p_list
        else :
            p_list_combined = np.concatenate([p_list_combined,p_list],axis=0)
    p_mean = np.mean(p_list_combined,axis=0)
    p_std = np.std(p_list_combined,axis=0)
    my_colors = cm.Oranges(index/n+0.3)
    plt.plot(np.arange(-1.5,1.5,0.05), p_mean, color=my_colors,label=f'$\sigma_1^2={r1_var},\sigma_2^2={TU-r1_var} $')
    plt.fill_between(np.arange(-1.5,1.5,0.05), p_mean + p_std, p_mean - p_std, color=my_colors, alpha=0.25)
plt.axvline(0, 0, 1, linestyle='--', color='black')
plt.axhline(0.5, 0, 1, linestyle='--', color='black')
plt.xlabel('value difference',size=20)
plt.ylabel('choice probability',size=20)
plt.legend(prop = {'size':12})
plt.xticks(size=20)
plt.yticks(size=20)
plt.show()

n = len(r1_var_list)
for (index,r1_var) in enumerate(r1_var_list):
    r1_std = np.sqrt(r1_var)
    r2_std = np.sqrt(TU-r1_var)
    p_list_combined = []
    for times in range(10):
        p_list = np.load(f'saved/simulation_TU={TU}_var={r1_var}_seeking_{times}_n=2.npy')
        p_list = p_list[np.newaxis,:]
        if times == 0:
            p_list_combined = p_list
        else :
            p_list_combined = np.concatenate([p_list_combined,p_list],axis=0)
    p_mean = np.mean(p_list_combined,axis=0)
    p_std = np.std(p_list_combined,axis=0)
    my_colors = cm.Blues(index/n+0.3)
    plt.plot(np.arange(-1.5,1.5,0.05), p_mean, color=my_colors,label=f'$\sigma_1^2={r1_var},\sigma_2^2={TU-r1_var} $')
    plt.fill_between(np.arange(-1.5,1.5,0.05), p_mean + p_std, p_mean - p_std, color=my_colors, alpha=0.25)
plt.axvline(0, 0, 1, linestyle='--', color='black')
plt.axhline(0.5, 0, 1, linestyle='--', color='black')
plt.xlabel('value difference',size=20)
plt.ylabel('choice probability',size=20)
plt.legend(prop = {'size':12})
plt.xticks(size=20)
plt.yticks(size=20)
plt.show()
```
