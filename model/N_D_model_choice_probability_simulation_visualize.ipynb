```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')
conds = ['averse', 'neutral', 'seeking']
for cond in conds:
    r_high = []
    r_middle = []
    r_low = []
    mean = []
    std = []
    total_n = 30
    for dim in range(2,11):
        r_list = np.load(f'saved/{dim}D_model_simulation_'+cond+'.npy')
        r_mean = np.mean(r_list,axis=0)
        r_std = np.std(r_list,axis=0)
        # print(r_list)
        mean.append(r_mean)
        std.append(r_std)
    mean = np.array(mean)
    std = np.array(std)
    plt.plot(np.arange(2, 11, 1), mean[:,2], label=f'Min($\sigma_i$)',marker='o',linestyle='-')
    plt.plot(np.arange(2,11,1),mean[:,0],label=f'Max($\sigma_i$)',marker='s',linestyle='-')
    plt.plot(np.arange(3, 11, 1), mean[1:,1], label=f'Median($\sigma_i$)',marker='v',linestyle='-')
    plt.plot(np.arange(2, 11, 1), [1 / 2, 1 / 3, 1 / 4, 1 / 5, 1 / 6, 1 / 7, 1 / 8, 1 / 9, 1 / 10], linestyle='--',
             color='black', label='1/n')
    plt.fill_between(np.arange(2,11,1),mean[:,0]+std[:,0]/np.sqrt(total_n),mean[:,0]-std[:,0]/np.sqrt(total_n),  color='#ff796c',alpha=0.25)
    plt.fill_between(np.arange(3, 11, 1), mean[1:, 1] + std[1:, 1]/np.sqrt(total_n), mean[1:, 1] - std[1:, 1]/np.sqrt(total_n), color='#75bbfd', alpha=0.25)
    plt.fill_between(np.arange(2, 11, 1), mean[:, 2] + std[:, 2]/np.sqrt(total_n), mean[:, 2] - std[:, 2]/np.sqrt(total_n), color='#06c2ac', alpha=0.25)
    plt.legend(prop={'size': 15})
    plt.xticks(size=15)
    plt.yticks(size=15)
    plt.ylim(0, 0.89)
    plt.xlabel('Dimensionality (N)', size=20)
    plt.ylabel('Probability', size=20)
    plt.show()
```
